```python
import hashlib
```

```python
string = input("Digite o texto a ser criptografado: ")
```

```python
menu = int(input(''' ## Menu - Escolha o tipo de Hash ## 

1) MD5
2) SHA1
3) SHA256 
4) SHA512

Digite o número do hash: '''))
```

```python
if menu == 1:
  resultado = hashlib.md5(string.encode('utf-8'))
  print("O hash da string ", string, " em MD5 é: ", resultado.hexdigest())
elif menu == 2:
  resultado = hashlib.sha1(string.encode('utf-8'))
  print("O hash da string ", string, " em SHA1 é: ", resultado.hexdigest())
elif menu == 3:
  resultado = hashlib.sha256(string.encode('utf-8'))
  print("O hash da string ", string, " em SHA256 é: ", resultado.hexdigest())
elif menu == 4:
  resultado = hashlib.sha512(string.encode('utf-8'))
  print("O hash da string ", string, " em SHA512 é: ", resultado.hexdigest())
else:
  print("O valor informado é inválido.")
```
